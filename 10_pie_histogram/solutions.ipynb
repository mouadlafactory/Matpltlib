```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("bootcamp_applicants_500.csv")

# Custom bins (for readability)
bins = [16, 18, 22, 26, 30, 35, 45]
labels = ["16-18 (HS)", "19-22 (Univ)", "23-26 (Early Career)", "27-30", "31-35", "36-45"]

df["age_range"] = pd.cut(df["age"], bins=bins, labels=labels, include_lowest=True)
df

```

```python
age_counts = df["age_range"].value_counts().sort_index()
age_counts
```

```python

plt.figure(figsize=(10,6))
bars = plt.bar(age_counts.index, age_counts.values, color="#4A90E2")

# Add labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 3,
             f"{height} ({height/len(df)*100:.1f}%)",
             ha="center", va="bottom", fontsize=9)

# Title & formatting
plt.title("Who is Applying? Age Distribution of GenAI & ML Bootcamp Applicants (n=500)", pad=20)
plt.xlabel("Age Groups")
plt.ylabel("Number of Applicants")
plt.grid(axis="y", alpha=0.25)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("bootcamp_applicants_500.csv")

# Count occurrences of each education level
edu_counts = df["education_level"].value_counts().reindex(["High school","Bachelor","Master","PhD"])
edu_counts
```

```python

plt.figure(figsize=(10,6))
bars = plt.bar(edu_counts.index, edu_counts.values, color="#00A676")  # greenish color for progress/education

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 3,
             f"{height} ({height/len(df)*100:.1f}%)",
             ha='center', va='bottom', fontsize=9)

# Title & labels
plt.title("What is the Education Level of GenAI/ML Applicants? (n=500)", pad=20)
plt.xlabel("Education Level")
plt.ylabel("Number of Applicants")
plt.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("bootcamp_applicants_500.csv")

gender_counts = df["gender"].value_counts()

plt.figure(figsize=(7,7))
colors = ["#1f77b4", "#ff69b4", "#8c564b"]  # blue (male), pink (female), brown (other)

plt.pie(
    gender_counts.values,
    labels=gender_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    explode=[0.03 if g=="Female" else 0 for g in gender_counts.index]  # highlight female slice
)

plt.title("Gender Distribution of GenAI/ML Applicants", pad=20)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("bootcamp_applicants_500.csv")

city_counts = df["city"].value_counts()

plt.figure(figsize=(8,8))

# Use a color map for more professional visualization
colors = plt.cm.Paired(range(len(city_counts)))

plt.pie(
    city_counts.values,
    labels=city_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors
)

plt.title("Where Are Applicants Coming From? (City Distribution)", pad=20)
plt.tight_layout()
plt.show()
```
