```python
import matplotlib.pyplot as plt

year = [2014, 2015, 2016, 2017]
income = [4000, 4500, 5300, 4600]
expense = [2800, 3000, 2800, 3400]

plt.figure(figsize=(10, 8))  # Set figure size

# Income subplot
plt.subplot(2, 1, 1) # 2 rows, 1 column, position 1 (top)
plt.bar(year, income, color='blue', alpha=0.7)
plt.title('Income Over Years')
plt.ylabel('Amount ($)')
plt.grid(True, alpha=0.3)

# Expense subplot
plt.subplot(2, 1, 2) # 2 rows, 1 column, position 2 (bottom)
plt.bar(year, expense, color='green', alpha=0.7)
plt.title('Expenses Over Years')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.grid(True, alpha=0.3)

plt.tight_layout()  # Prevents overlap
plt.show()
```
