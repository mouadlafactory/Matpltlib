Company Revenue Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df_revenue = pd.read_csv("revenue_100.csv")

# Show first rows
print(df_revenue.head())
```

Convert date column to datetime

```python
df_revenue['date'] = pd.to_datetime(df_revenue['date'])
```

```python
plt.figure(figsize=(12, 6))  # wider chart
plt.plot(df_revenue['date'], df_revenue['revenue'])
plt.title("Company Revenue Trend (100 Days)")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_revenue['date'],
         df_revenue['revenue'],
         color='blue',
         linewidth=2,
         marker='o',
         markersize=4)

plt.title("Company Revenue Trend (with Styling)", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.grid(alpha=0.3)             # soft grid
plt.tight_layout()              # better layout
plt.show()
```

```python
start_value = df_revenue['revenue'].iloc[0]
end_value = df_revenue['revenue'].iloc[-1]
growth = end_value - start_value

print(f"Starting revenue: {start_value}")
print(f"Ending revenue: {end_value}")
print(f"Growth over 100 days: {growth:.2f}")
```
