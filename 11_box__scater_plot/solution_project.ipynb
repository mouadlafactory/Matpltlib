````python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/Users/newuser/Desktop/gen-ai/gen_ai_machine_learning_bootcamp/gen_ai_machine_learning/py/matpltlib/11_box__scater_plot/climate_energy_morocco_5years.csv")

# Climate colors (matching seaborn version)
colors = {
    "North": "#1f77b4",   # blue/cool
    "Center": "#2ca02c",  # green/mild
    "South": "#ff7f0e"    # orange/hot
}

seasons = ["Winter", "Spring", "Summer", "Autumn"]
regions = ["North", "Center", "South"]

fig, ax = plt.subplots(figsize=(14,10))

# Collect boxplot data by season + region
data = []
color_list = []
positions = []
pos = 1

for season in seasons:
    for region in regions:
        values = df[(df["season"] == season) & (df["region"] == region)]["energy_kwh"]
        data.append(values)
        color_list.append(colors[region])
        positions.append(pos)
        pos += 1
    pos += 1  # spacing between seasons

# Draw boxplots
bp = ax.boxplot(data, positions=positions, patch_artist=True, widths=0.6)

# Apply colors
for patch, color in zip(bp['boxes'], color_list):
    patch.set_facecolor(color)
    patch.set_alpha(0.8)

# X-axis seasonal labels
xticks_positions = [ (positions[i] + positions[i+2]) / 2 for i in range(0, len(positions), 3) ]
print(xticks_positions)
ax.set_xticks(xticks_positions)
ax.set_xticklabels(seasons)

# Title and Labels
ax.set_title("Energy Consumption by Season and Region (5-year Climate Pattern)", pad=20)
ax.set_xlabel("Season")
ax.set_ylabel("Daily Energy Consumption (kWh)")

# Grid and layout
ax.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()

````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/Users/newuser/Desktop/gen-ai/gen_ai_machine_learning_bootcamp/gen_ai_machine_learning/py/matpltlib/11_box__scater_plot/climate_energy_morocco_5years.csv")

# Climate colors (region-based)
colors = {"North": "#1f77b4", "Center": "#2ca02c", "South": "#ff7f0e"}

plt.figure(figsize=(12,8))

# Scatter by region
for region, color in colors.items():
    subset = df[df["region"] == region]
    plt.scatter(
        subset["temperature_c"], 
        subset["energy_kwh"], 
        alpha=0.20,          # light transparency so dense areas show darker
        s=10,                # small point size for large dataset
        color=color,
        label=region
    )

plt.title("Temperature vs Energy Consumption — Scatter by Region", pad=20)
plt.xlabel("Temperature (°C)")
plt.ylabel("Daily Energy Consumption (kWh)")
plt.grid(alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
````

| Observation                                             | Explanation                     |
| ------------------------------------------------------- | ------------------------------- |
| Dots are LOWEST around 20–22°C                          | Energy comfort zone             |
| More spread (taller cloud) above 28°C                   | AC / cooling stress             |
| More spread below ~15°C                                 | Heating stress                  |
| South cluster is higher on average                      | Desert/hot regions consume more |
| North cluster is lower on right side but higher on left | Heating > cooling               |

````python
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)

data = [data1, data2, data3]

fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)

# Creating axes instance and the box plot
bp = ax.boxplot(data, patch_artist = True, notch = True, vert = 0)

# Changing color and line width of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color ='blue', linewidth = 1.5, linestyle =":")

# Changing color and line width of the caps
for cap in bp['caps']:
    cap.set(color ='red', linewidth = 2)

# Changing color and linewidth of the medians
for median in bp['medians']:
    median.set(color ='green', linewidth = 3)

# Changing style of fliers
for flier in bp['fliers']:
    flier.set(marker ='D', color ='#e7298a', alpha = 0.5)

# Adding title 
plt.title('Customized Box Plot')
plt.show()
````



1. Generates **3 artificial datasets** using `numpy`
2. Puts them into a **boxplot**
3. Customizes colors and styles for whiskers, caps, medians, and outliers
4. Displays a **horizontal (vert=0) boxplot**
5. Makes it visually attractive and more readable

---

## 1️⃣ Creating Sample Data

```python
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)
```

Each of these lines uses `np.random.normal(mean, std, size)` to generate synthetic data.

| Variable | Mean | Std Dev | Size       | Interpretation                    |
| -------- | ---- | ------- | ---------- | --------------------------------- |
| `data1`  | 100  | 10      | 200 points | Values cluster tightly around 100 |
| `data2`  | 90   | 20      | 200 points | Wider spread than data1           |
| `data3`  | 80   | 30      | 200 points | Very spread out, more variability |

This simulates **three populations with different means and variances**.

Then we combine them:

```python
data = [data1, data2, data3]
```

Now `data` is a **list of 3 datasets**, perfect for a boxplot with 3 boxes.

---

## 2️⃣ Creating the Figure

```python
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
```

* `plt.figure` → creates the canvas
* `figsize` → width = 10, height = 7
* `add_subplot(111)` → 1 row, 1 column, first subplot = full figure

---

## 3️⃣ Creating the Boxplot

```python
bp = ax.boxplot(data, patch_artist=True, notch=True, vert=0)
```

| Argument            | Meaning                                              |
| ------------------- | ---------------------------------------------------- |
| `data`              | list of arrays to plot                               |
| `patch_artist=True` | allows coloring the box (filled color)               |
| `notch=True`        | adds a notch around the median for better visibility |
| `vert=0`            | **horizontal boxplot** instead of vertical           |

A **horizontal boxplot** is easier to read when comparing distributions.

---

## 4️⃣ Customizing the Boxplot

Now we modify its appearance:

### ✅ Whiskers

```python
for whisker in bp['whiskers']:
    whisker.set(color ='blue', linewidth = 1.5, linestyle =":")
```

Whiskers show the **range** of non-outlier data.
We change:

* color = blue
* style = dotted (`:`)
* width = 1.5

---

### ✅ Caps

```python
for cap in bp['caps']:
    cap.set(color ='red', linewidth = 2)
```

Caps are the **little horizontal lines** at the ends of whiskers.

---

### ✅ Medians

```python
for median in bp['medians']:
    median.set(color ='green', linewidth = 3)
```

The median (50th percentile) is now a thick **green** line — easy to spot.

---

### ✅ Outliers (Fliers)

```python
for flier in bp['fliers']:
    flier.set(marker ='D', color ='#e7298a', alpha = 0.5)
```

* Uses a **diamond marker** (`D`)
* Pink color
* Semi-transparent (`alpha=0.5`)
  → Makes outliers visible and stylish

---

## 5️⃣ Show the Plot

```python
plt.title('Customized Box Plot')
plt.show()
```

---

# ✅ Final Interpretation

| Data  | Mean               | Spread             | Visualization effect |
| ----- | ------------------ | ------------------ | -------------------- |
| data1 | Highest mean (100) | smallest spread    | box is narrow        |
| data2 | lower mean (90)    | medium spread      | box is wider         |
| data3 | lowest mean (80)   | **largest spread** | box is very wide     |

This demonstrates **how variance affects box width**.
